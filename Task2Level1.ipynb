```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
print("Step 1: Creating a sample sales dataset...")

data = {
    'Date': pd.to_datetime([
        '2023-01-05', '2023-01-05', '2023-01-06', '2023-01-06', '2023-01-07',
        '2023-01-07', '2023-01-08', '2023-01-08', '2023-01-09', '2023-01-09'
    ]),
    'Region': [
        'North', 'South', 'North', 'East', 'South',
        'West', 'East', 'West', 'North', 'South'
    ],
    'Product': [
        'Laptop', 'Keyboard', 'Mouse', 'Laptop', 'Monitor',
        'Keyboard', 'Mouse', 'Monitor', 'Laptop', 'Keyboard'
    ],
    'UnitsSold': [10, 25, 30, 8, 15, 20, 28, 12, 11, 22],
    'Price': [1200, 75, 25, 1250, 300, 80, 27, 320, 1150, 78]
}
df = pd.DataFrame(data)
df['TotalSales'] = df['UnitsSold'] * df['Price']
print("Sample dataset created successfully.\n")
print("Original Data:")
print(df)
print("\n" + "="*50 + "\n")
```

```python
print("Aggregating data by Region...")
region_summary = df.groupby('Region').agg(
    TotalSales=('TotalSales', 'sum'),
    AverageSales=('TotalSales', 'mean'),
    TotalUnitsSold=('UnitsSold', 'sum')
).reset_index()

print("\nAggregation by Region:")
print(region_summary)
```

```python
print("\nAggregating data by Product...")
product_summary = df.groupby('Product').agg(
    TotalUnitsSold=('UnitsSold', 'sum'),
    AveragePrice=('Price', 'mean')
).reset_index()

print("\nAggregation by Product:")
print(product_summary)
```

```python
print("\nAggregating data by Date...")
daily_sales = df.groupby('Date')['TotalSales'].sum().reset_index()

print("\nAggregation by Date (Daily Sales):")
print(daily_sales)
print("\nAggregation complete.\n")
print("\n" + "="*50 + "\n")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=region_summary, x='Region', y='TotalSales', palette='viridis')
plt.title('Total Sales by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('sales_by_region.png')
print("Saved sales_by_region.png")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=product_summary, x='Product', y='TotalUnitsSold', palette='plasma')
plt.title('Total Units Sold by Product', fontsize=16)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Total Units Sold', fontsize=12)
plt.tight_layout()
plt.savefig('units_by_product.png')
print("Saved units_by_product.png")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_sales, x='Date', y='TotalSales', marker='o', color='teal')
plt.title('Daily Sales Trend', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('daily_sales_trend.png')
print("Saved daily_sales_trend.png")

print("\nVisualizations generated. Script finished.")
```
