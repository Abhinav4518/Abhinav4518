A dataset with an unequal class distribution


```python
import numpy as np
import pandas as pd
```

```python
credit_card_data=pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset.

```python
#0->legit transcation
#1->fraudlent transcation
#seperating the legit and fraudlent transcations
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Under-Sampling

```python
#Building a sample dataset containing similar distribution of Legit & Fraudlent Transcation
```

```python
legit_sample=legit.sample(n=102)
```

```python
print(legit_sample.shape)
```

```python
#Concatenate the two colums
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
