**Methods to handle missing value**
1.Imputation
2.Dropping

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
dataset =pd.read_csv('/content/School_Attendance_by_Student_Group_and_District__2021-2022.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

Central Tendencies:
1. Mean
2. Median
3. Mode


```python
#analyse the distribution of data in the salary
fig,ax=plt.subplots(figsize=(8,8))
sns.histplot(dataset.Category)
```

Replace the missing values with median value


```python
dataset['2019-2020 student count'].fillna(dataset['2019-2020 student count'].median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
#filling with mean values
dataset['2020-2021 student count'].fillna(dataset['2020-2021 student count'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```
