Label Encoding:
1. Converting the labels into numeric form


```python
#importing library
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding Breast Cancer Dataset

```python
#loading the data from csv file to pandas dataframe
cancer_data=pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
#find the count of different status
cancer_data['Status'].value_counts()
```

```python
#load the label Encoder function
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.Status)
```

```python
#appending the labels to the DataFrame
cancer_data['target']=labels
cancer_data.head()
```

```python
#0->Alive 1->Dead
cancer_data['target'].value_counts()
```
